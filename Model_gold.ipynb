### Проверим формулу вычисления эффективности обогащения

Эффективность обогащения рассчитывается по формуле: Recovery =C*(F-T)/F*(C-T)*100%, где:   
С - доля золота в концентрате после флотации/очистки,   
F - доля золота в сырье/концентрате до флотации/очистки,   
Т - доля золота в отвальных хвостах после флотации/очистки.

```python
C = df_train['rougher.output.concentrate_au']
F = df_train['rougher.input.feed_au']
T = df_train['rougher.output.tail_au']
```

Создадим новый столбец для проведения расчетов по проверке эффективности обогащения.

```python
df_train['control_recovery'] = ((C*(F-T))/(F*(C-T)))*100
df_train['control_recovery'].mean()
```

Получим значения по признаку rougher.output.recovery и сравним результаты.


```python
df_train['rougher.output.recovery'].mean()
```

Значения совпадают. Для большей вероятности построим график и сравним визуально.

```python
df_train['control_recovery'].hist()
plt.xlabel('Recovery - результаты эффективности обогащения')
plt.ylabel('Количество наблюдений')
plt.suptitle('Распределение результатов эффективности обогащения (расчетный результат)', fontsize=20)
plt.show()
```

```python
df_train['rougher.output.recovery'].hist()
plt.xlabel('Recovery - результаты эффективности обогащения')
plt.ylabel('Количество наблюдений')
plt.suptitle('Распределение результатов эффективности обогащения (исходные данные)', fontsize=20)
plt.show()
```

Графики идентичны. В расчете вычисления эффективности обогащения визуально разночтений не выявлено.

Найдем среднюю абсолютную ошибку (MAE) между нашими расчётами и значением признака.

### Напишем функцию для вычисления итоговой sMAPE

Для начала напишем функцию sMAPE.

```python
def smape_total(target_rougher, predicted_rougher, target_final, predicted_final):
    smape_rough = 1/len(target_rougher) * np.sum(2 * np.abs(
      target_rougher - predicted_rougher) / (np.abs(target_rougher) + np.abs(predicted_rougher))*100)
    smape_fin = 1/len(target_final) * np.sum(2 * np.abs(
      target_final - predicted_final) / (np.abs(target_final) + np.abs(predicted_final))*100)
    return (0.25*smape_rough) + (0.75*smape_fin)
```

```python
def smape(target, predicted):
    return 1/len(target) * np.sum(2 * np.abs(
       target - predicted) / (np.abs(target) + np.abs(predicted))*100)
def smape_final(rougher, final):
    return 0.25*rougher + 0.75*final
```
